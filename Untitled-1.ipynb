```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Data Preprocessing

```python
# a. load dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
df.head()
```

```python
# b. split Blood Pressure into Systolic & Diastolic to make a Blood Pressure Category
df[['systolic', 'diastolic']] = (df['Blood Pressure']
                                 .str.split('/', expand=True)
                                 .astype(int)
                                )
df.head()
```

```python
# Chek NaN values
nan_colums = df.columns[df.isnull().any()]
nan_count = df.isnull().sum()

if len(nan_colums) > 0:
    print("Column that have NaN values:")
    for column in nan_colums:
        print(f"{column}: {nan_count[column]} NaN values")
        
    nan_rows = df[df.isnull().any(axis=1)]

    # Show rows that contain NaN values
    print("Rows with NaN values:", nan_rows)

    # If you want just the indices of these rows
    nan_row_indices = nan_rows.index.tolist()
    print("Indices of rows with NaN values:", nan_row_indices)
else:
    print("NaN values not found")
```

```python
# function to category the blood pressure
def bp_category(diastolic, systolic) -> str:
    if systolic < 90 or diastolic < 60:
        return "low"
    elif (90 <= systolic < 120) and (60 <= diastolic < 80):
        return "normal"
    elif (120 <= systolic < 140) or (80 <= diastolic < 90):
        return "pre-high"
    elif systolic >= 140 or diastolic >= 90:
        return "high"

# function to call the bp_category funtion
def apply_bp_category(row):
    return bp_category(row['diastolic'], row['systolic'])

```

```python
# function to call the appky_bp_category funtion
df['bp_category'] = df.apply(apply_bp_category, axis=1)
df.head()
```

```python
df.info()
```

```python
# Drop the columns that has no significant impact and lacks direct relevance to the diagnosis of the patientâ€™s sleep disorder
df = df.drop('Person ID', axis = 1)
df = df.drop('Blood Pressure', axis = 1)
df = df.drop('systolic', axis = 1)
df = df.drop('diastolic', axis = 1)
df.head()
```

```python
# Function for encode
def encoding_time(colums):
    mapping_thing = {}

    encoded_datas = []
    counter = 0

    for c in colums:
        if c not in mapping_thing:
            mapping_thing[c] = counter
            counter += 1

        # adding key(number) & value (value of column)
        encoded_data = mapping_thing[c]

        # add the kiri to the array encoded_data
        encoded_datas.append(encoded_data)
        
    return encoded_datas
```

```python
# Call the encoding_time funtion & replace the value inside the column
array_Gender = df['Gender'].unique().tolist()

print(array_Gender)
encode_Gender = encoding_time(array_Gender)
print(encode_Gender)
df['Gender'] = df['Gender'].replace(array_Gender, encode_Gender)

array_Occupation = df['Occupation'].unique().tolist()
encode_Occupation = encoding_time(array_Occupation)
print(array_Occupation)
print(encode_Occupation)
df['Occupation'] = df['Occupation'].replace(array_Occupation, encode_Occupation)

array_BMICategory = df['BMI Category'].unique().tolist()
print(array_BMICategory)
encode_BMICategory = encoding_time(array_BMICategory)
print(encode_BMICategory)
df['BMI Category'] = df['BMI Category'].replace(array_BMICategory, encode_BMICategory)

array_SleepDisorder = df['Sleep Disorder'].unique().tolist()
encode_SleepDisorder = encoding_time(array_SleepDisorder)
print(array_SleepDisorder)
print(encode_SleepDisorder)
df['Sleep Disorder'] = df['Sleep Disorder'].replace(array_SleepDisorder, encode_SleepDisorder)

# When using the function, it only 'slightly raised', 'high', and 'normal. So, i encode from the dictionary of bp_category and then replace the value
array_bp_category = df['bp_category'].unique().tolist()
print(array_bp_category)
df['bp_category'].replace({'low':0, 'normal': 1, 'pre-high': 2, 'high': 3}, inplace=True)
array_bp_category = df['bp_category'].unique().tolist()
print(array_bp_category)
```

```python
# Assign the predictor and target variables
X = df.loc[:, df.columns != 'Sleep Disorder'].values
y = df['Sleep Disorder'].values
# print(X)
```

```python
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=17)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

## 3. Model Generation

```python
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.simplefilter(action="ignore")

rf_model = RandomForestClassifier(criterion="gini", n_estimators=200, max_depth=20, min_samples_split=2, min_samples_leaf=5, max_features="sqrt", bootstrap=True, class_weight="balanced", random_state=1, oob_score=True, n_jobs=1)
# rf_model = RandomForestClassifier(n_estimators = 100, min_samples_split = 5, max_depth= 5,  criterion = 'gini', random_state = 0)
# rf_model = RandomForestClassifier(max_depth=5, random_state=3)
# rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
rf_model
```

## 4. Model Evaluation

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, classification_report, confusion_matrix

cm = confusion_matrix(y_test,y_pred)

print("Confusion Matrix :\n", cm,"\n")


print(classification_report(y_test, y_pred))

print(f"Accuracy: {round(accuracy_score(y_test, y_pred), 4)}")

# print(f"Recall: {round(recall_score(y_test, y_pred, average='micro'), 4)}")

# print(f"Precision: {round(precision_score(y_test, y_pred, average='micro'), 4)}")

# print(f"F1: {round(f1_score(y_test, y_pred, average='micro'), 4)}")

# print(f"Auc: {round(roc_auc_score(y_pred, y_test, multi_class='ovr'), 4)}")
```

```python

print(f"Recall: {round(recall_score(y_pred, y_test, average='weighted'), 4)}")

print(f"Precision: {round(precision_score(y_pred, y_test, average='weighted'), 4)}")

print(f"F1: {round(f1_score(y_pred, y_test, average='weighted'), 4)}")
```
